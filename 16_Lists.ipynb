```python
#ë¦¬ìŠ¤íŠ¸ ìš”ì†Œ ë¬´ì‘ìœ„ë¡œ ì„ê¸°

from random import shuffle

listdata = list(range(1,11))
for i in range(3): # ì¦‰, 0, 1, 2 ì„¸ ë²ˆ ë°˜ë³µí•˜ë¼ëŠ” ëœ»
  shuffle(listdata)
  print(listdata)


#ë¦¬ìŠ¤íŠ¸ ìš”ì†Œ ë¬´ì‘ìœ„ë¡œ ì„ê¸°ë¥¼ ë‚´ê°€ ì„ì˜ë¡œ ë³€í˜• ì‹œì¼œë³´ë‹ˆê¹Œ ê²°ê³¼ê°€ íŒŒì´ì¬200ì œì™€ ë‹¤ë¦„. ë‚˜ëŠ” ì´ í•¨ìˆ˜ê°€ ê°ì²´ë¥¼ ë¦¬í„´í•˜ëŠ” ê²ƒì— ëŒ€í•´ ë” ì •í™•íˆ ì´í•´í•  í•„ìš”ê°€ ìˆìŒ.
listdata = list(range(1,11))
for i in range(3): # ì¦‰, 0, 1, 2 ì„¸ ë²ˆ ë°˜ë³µí•˜ë¼ëŠ” ëœ»
  ret = shuffle(listdata)
  print(ret)

else:
  print(listdata)

```
