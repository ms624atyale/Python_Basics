# Digital Signal Processing
> sampling rate(Hz), frequency(Hz), phasor

```python
#@markdown ğŸ‘€ ğŸ¾ <font color = 'red'> **Step 0: Install numpy, matplotlib, and IPython packages**

import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

## ğŸ”‰ **Phasor (sine)**

```python
#@markdown ğŸ‘€ ğŸ¾ <font color = 'red'> **1. Parameter setting** â¤µï¸ 

#@markdown <font color = 'blue'> **amplitude, sampling rate, duration, frequency with 200Hz, & phase**

amp = 1          # range [0.0, 1.0]
sr = 10000       # sampling rate, Hz
dur = 0.5        # in seconds
freq = 200.0     # sine frequency, Hz
phase = 0  # phase, radian
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color = 'red'> **2. Generate sample ticks** â¤µï¸ 

nSamp = sr * dur
samples = np.arange(1, nSamp+1)
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color = 'red'> **3. Convert sample to time (sec)** â¤µï¸ 

t = samples/sr
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color = 'red'> **4. Convert time to angle (radian)** â¤µï¸ 

theta0 = t * 2*np.pi * freq
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color = 'red'> **5. Shift angle by phase** â¤µï¸ 

theta = theta0 + phase
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color = 'red'> **6. Generate signal by sin-phasor** â¤µï¸ 

s = amp * np.sin(theta)
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color = 'red'> **7. Plot the entire signal** â¤µï¸ 

plt.plot(theta, s, ':')
plt.xlabel('angle (radian)')
plt.ylabel('real')
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color = 'red'> **8. Plot a certain portion of signals** â¤µï¸ 
sampleRng = np.arange(0,100)
plt.plot(t[sampleRng], s[sampleRng], ':')
plt.xlabel('time (s)')
plt.ylabel('real')
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color = 'red'> **9. Write and generate a waveform with 100Hz of fundamental frequency** â¤µï¸ 

ipd.Audio(s, rate=sr)
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color = 'red'> **10. PHASOR_SINE ( ) function defined** â¤µï¸ 

def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color = 'red'> **11. Write and generate a waveform with 110Hz of f0** â¤µï¸ 
amp = 1; sr = 10000; dur = 1; freq = 200; phase = 0
s, t, theta = phasor_sine(amp, sr, dur, freq, phase)
ipd.Audio(s, rate=sr)
```

## ğŸ”‰ **Phasor (complex)**

```python
#@markdown ğŸ‘€ ğŸ¾ <font color ='blue'> **1. Parameter setting** â¤µï¸

amp = 1          # range [0.0, 1.0]
sr = 10000       # sampling rate, Hz
dur = 0.5        # in seconds
freq = 50.0     # sine frequency, Hz
phase = 0  # phase, radian
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color ='blue'> **2. Generate sample ticks** â¤µï¸

nSamp = sr * dur
samples = np.arange(1, nSamp+1)
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color ='blue'> **3. Convert samples to time** â¤µï¸

t = samples/sr
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color ='blue'> **4. Convert time to angle in radian** â¤µï¸

theta0 = t * 2*np.pi * freq
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color ='blue'> **5. Shift angle by phase** â¤µï¸

theta = theta0 + phase
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color ='blue'> **6. Generate signal by complex-phasor** â¤µï¸

c = amp * np.exp(theta*1j)
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color ='blue'> **7. Plot the entire signal** â¤µï¸

import plotly.graph_objects as go
x, y, z = t, c.real, c.imag
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z,
                                   mode='markers',
                                   marker=dict(size=3))])
fig.update_layout(scene = dict(
                    xaxis_title='time (sec)',
                    yaxis_title='real',
                    zaxis_title='imag'))
fig.show()
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color ='blue'> **8. Plot a certain portion of signals** â¤µï¸
import plotly.graph_objects as go
sampleRng = np.arange(0,100)
x, y, z = t[sampleRng], c.real[sampleRng], c.imag[sampleRng]
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z,
                                   mode='markers',
                                   marker=dict(size=3))])
fig.update_layout(scene = dict(
                    xaxis_title='time (sec)',
                    yaxis_title='real',
                    zaxis_title='imag'))
fig.show()
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color ='blue'> **9. Write and generate 100Hz of Fundamental frequency** â¤µï¸

ipd.Audio(c, rate=sr)
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color ='blue'> **10. PHASOR_COMPLEX ( ) function defined** â¤µï¸

def phasor_complex(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  c = amp * np.exp(theta*1j)
  return c, t, theta
```

```python
#@markdown ğŸ‘€ ğŸ¾ <font color ='blue'> **11. Write and generate 500Hz of Fundamental frequency** â¤µï¸

amp = 1; sr = 10000; dur = 1; freq = 100; phase = 0
c, t, theta = phasor_complex(amp, sr, dur, freq, phase)
ipd.Audio(c.real, rate=sr)
```
