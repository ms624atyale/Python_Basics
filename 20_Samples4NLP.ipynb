```python
#@markdown ðŸŽ¯ðŸ“Œ ë¬¸ìžì—´ì˜ ê° ë¬¸ìžë¥¼ ê·¸ ë‹¤ìŒ ë¬¸ìžë¡œ ë³€ê²½í•˜ê¸°

seq = input('ë‹¨ì–´, êµ¬ë¬¸, í˜¹ì€ ë¬¸ìž¥ì„ ìž…ë ¥í•˜ì„¸ìš”:')

ret = ''            #missed! remember!
for i in range(len(seq)):
  if i !=len(seq)-1: #textì˜ ê° ë¬¸ìž index iì— ëŒ€í•˜ì—¬ iê°€ textì˜ ë§¨ ë§ˆì§€ë§‰ ë¬¸ìžì˜ ì¸ë±ìŠ¤ê°€ ì•„ë‹ˆë©´ (ì´ ë•Œ -1 ì´ raw ë¡œ ì‚¬ìš©ë¨. [] ì•ˆì— ìžˆì§€ ì•ŠìŒ. ì´í•´X) 
    ret += seq[i+1] #retì— txt[i+1]ì„ ì¶”ê°€í•œë‹¤.
  else:             #index iê°€ txtì˜ ë§ˆì§€ë§‰ ë¬¸ìžë¼ë©´
    ret +=seq[0]    #retì— txt[0]ì„ ì¶”ê°€í•œë‹¤.  

print(ret)
```

```python
#@markdown ðŸŽ¯ðŸ“Œ URLì„ slash'/'ë¡œ êµ¬ë¶„í•œ ë¬¸ìžì—´ì´ ìš”ì†Œì¸ ë¦¬ìŠ¤íŠ¸ë¡œ ë§Œë“¤ì–´ URLì—ì„œ ì¿¼ë¦¬ ë¬¸ìžì—´ ì¶”ì¶œí•˜ê¸°
#ì¿¼ë¦¬ ë¬¸ìžì—´ì€ '?' ë’¤ì— í‘œì‹œë˜ëŠ” ë¬¸ìžì—´ì„ ì§€ì¹­í•˜ëŠ”ë°, ì¿¼ë¦¬ ë¬¸ìžì—´ì€ ë³€ìˆ˜=ê°’ ìŒì´ &ë¡œ êµ¬ë¶„ë˜ì–´ ë‚˜ì—´ëœë‹¤. 

url = 'https://www.google.com/search?q=google+translate&oq=google+tran&aqs=chrome.0.69i59i512j69i57j0i512l4j0i10i512j0i512l3.3294j0j4&sourceid=chrome&ie=UTF-8'
temp = url.split('?') #ë³€ìˆ˜ tempì— urlì„ '?'ë¡œ êµ¬ë¶„í•œ ë¬¸ìžì—´ì´ ìš”ì†Œì¸ ë¦¬ìŠ¤íŠ¸ë¥¼ í• ë‹¹í•˜ë¼.
queries = temp[1].split('&') #temp[1]ì€ ? ë‹¤ìŒì— ë‚˜ì˜¤ëŠ” ë¬¸ìžì—´. ì´ ë¬¸ìžì—´ì„ '&'ë¡œ êµ¬ë¶„í•œ ê²°ê³¼ë¥¼ query (a pair of variable=value)ë¡œ ë§Œë“ ë‹¤.

for query in queries:
  print(query)


```

```python
#@markdown ðŸŽ¯ðŸ“Œ ë¬¸ìž¥ì— ë‚˜íƒ€ë‚˜ëŠ” ë¬¸ìž ë¹ˆë„ìˆ˜ ê³„ì‚°í•˜ê¸°

def getTextfreq(filename):
  with open(filename,'rt') as f:
    text = f.read()
    f_dic = {} #ë¹ˆ ì‚¬ì „ìœ¼ë¡œ ë‘” f_dic ëŠ” ì¶”í›„ í•¨ìˆ˜ê°€ ì ìš©ëœ ë•Œì—ëŠ” ë¬¸ìž(key):ë¹ˆë„ìˆ˜(value)ê°€ í•œìŒìœ¼ë¡œ ì´ë£¨ì–´ì§„ ìš”ì†Œì¸ ì‚¬ì „ìœ¼ë¡œ êµ¬ì„±ë  ê²ƒì´ë‹¤.  
    for c in text: #í…ìŠ¤íŠ¸ì˜ ëª¨ë“  ë¬¸ìžë¥¼ í•œ ê°œì”© ë„ì§‘ì–´ë‚´ì„œ
      if c in f_dic: #ë§Œì•½ ë¬¸ìžê°€ ì‚¬ì „ f_dicì— key ë¡œ ì¡´ìž¬í•œë‹¤ë©´
        f_dic[c]+=1 # í•´ë‹¹ê°’ì„ 1 ì”© ì¦ê°€ì‹œì¼œ ì¤€ë‹¤.
      else:         #ë§Œì•½ ë¬¸ìžê°€ ì‚¬ì „ f_dicì— key ë¡œ ì¡´ìž¬í•˜ì§€ ì•ŠëŠ”ë‹¤ë©´
        f_dic[c]=1  #ìƒˆë¡œìš´ ìš”ì†Œ {c:1} ìŒì„ f_dicì— ì¶”ê°€í•œë‹¤. 
  return f_dic

ret1 = getTextfreq('/content/sample_data/exercise/crime_punish_ch_one.txt')
ret2 = sorted(ret1.items(), key=lambda x:x[1], reverse=True) #ret1ì„ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•˜ì—¬ ret2ë¡œ ìž¬ì§€ì •. ret2ëŠ” getTextfreq()í•¨ìˆ˜ê°€ ë¦¬í„´í•œ dictionary!!!! ì‚¬ì „ ìžë£Œ ì •ë¦¬ëŠ” refer to #131.
for c, freq in ret2:
  if c =='\n': #ë¬¸ìžê°€ ì¤„ë°”ê¿ˆ ê¸°í˜¸('\n')ì´ë©´
    continue  #í™”ë©´ì— ì¶œë ¥í•˜ì§€ ì•Šê³  ê·¸ëŒ€ë¡œ ë„˜ì–´ê°„ë‹¤.
  print('[%c] -> [%d]íšŒ ë‚˜íƒ€ë‚¨' %(c,freq))
```
