```python
#@markdown ğŸ¯ğŸ“Œ ë¬¸ìì—´ì˜ ê° ë¬¸ìë¥¼ ê·¸ ë‹¤ìŒ ë¬¸ìë¡œ ë³€ê²½í•˜ê¸°

seq = input('ë‹¨ì–´, êµ¬ë¬¸, í˜¹ì€ ë¬¸ì¥ì„ ì…ë ¥í•˜ì„¸ìš”:')

ret = ''            #missed! remember!
for i in range(len(seq)):
  if i !=len(seq)-1: #textì˜ ê° ë¬¸ì index iì— ëŒ€í•˜ì—¬ iê°€ textì˜ ë§¨ ë§ˆì§€ë§‰ ë¬¸ìì˜ ì¸ë±ìŠ¤ê°€ ì•„ë‹ˆë©´ (ì´ ë•Œ -1 ì´ raw ë¡œ ì‚¬ìš©ë¨. [] ì•ˆì— ìˆì§€ ì•ŠìŒ. ì´í•´X) 
    ret += seq[i+1] #retì— txt[i+1]ì„ ì¶”ê°€í•œë‹¤.
  else:             #index iê°€ txtì˜ ë§ˆì§€ë§‰ ë¬¸ìë¼ë©´
    ret +=seq[0]    #retì— txt[0]ì„ ì¶”ê°€í•œë‹¤.  

print(ret)
```

```python
#@markdown ğŸ¯ğŸ“Œ URLì„ slash'/'ë¡œ êµ¬ë¶„í•œ ë¬¸ìì—´ì´ ìš”ì†Œì¸ ë¦¬ìŠ¤íŠ¸ë¡œ ë§Œë“¤ì–´ URLì—ì„œ ì¿¼ë¦¬ ë¬¸ìì—´ ì¶”ì¶œí•˜ê¸°
#ì¿¼ë¦¬ ë¬¸ìì—´ì€ '?' ë’¤ì— í‘œì‹œë˜ëŠ” ë¬¸ìì—´ì„ ì§€ì¹­í•˜ëŠ”ë°, ì¿¼ë¦¬ ë¬¸ìì—´ì€ ë³€ìˆ˜=ê°’ ìŒì´ &ë¡œ êµ¬ë¶„ë˜ì–´ ë‚˜ì—´ëœë‹¤. 

url = 'https://www.google.com/search?q=google+translate&oq=google+tran&aqs=chrome.0.69i59i512j69i57j0i512l4j0i10i512j0i512l3.3294j0j4&sourceid=chrome&ie=UTF-8'
temp = url.split('?') #ë³€ìˆ˜ tempì— urlì„ '?'ë¡œ êµ¬ë¶„í•œ ë¬¸ìì—´ì´ ìš”ì†Œì¸ ë¦¬ìŠ¤íŠ¸ë¥¼ í• ë‹¹í•˜ë¼.
queries = temp[1].split('&') #temp[1]ì€ ? ë‹¤ìŒì— ë‚˜ì˜¤ëŠ” ë¬¸ìì—´. ì´ ë¬¸ìì—´ì„ '&'ë¡œ êµ¬ë¶„í•œ ê²°ê³¼ë¥¼ query (a pair of variable=value)ë¡œ ë§Œë“ ë‹¤.

for query in queries:
  print(query)


```

```python
#@markdown ğŸ¯ğŸ“Œ ë¬¸ì¥ì— ë‚˜íƒ€ë‚˜ëŠ” ë¬¸ì ë¹ˆë„ìˆ˜ ê³„ì‚°í•˜ê¸°

def getTextfreq(filename):
  with open(filename,'rt') as f:
    text = f.read()
    f_dic = {} #ë¹ˆ ì‚¬ì „ìœ¼ë¡œ ë‘” f_dic ëŠ” ì¶”í›„ í•¨ìˆ˜ê°€ ì ìš©ëœ ë•Œì—ëŠ” ë¬¸ì(key):ë¹ˆë„ìˆ˜(value)ê°€ í•œìŒìœ¼ë¡œ ì´ë£¨ì–´ì§„ ìš”ì†Œì¸ ì‚¬ì „ìœ¼ë¡œ êµ¬ì„±ë  ê²ƒì´ë‹¤.  
    for c in text: #í…ìŠ¤íŠ¸ì˜ ëª¨ë“  ë¬¸ìë¥¼ í•œ ê°œì”© ë„ì§‘ì–´ë‚´ì„œ
      if c in f_dic: #ë§Œì•½ ë¬¸ìê°€ ì‚¬ì „ f_dicì— key ë¡œ ì¡´ì¬í•œë‹¤ë©´
        f_dic[c]+=1 # í•´ë‹¹ê°’ì„ 1 ì”© ì¦ê°€ì‹œì¼œ ì¤€ë‹¤.
      else:         #ë§Œì•½ ë¬¸ìê°€ ì‚¬ì „ f_dicì— key ë¡œ ì¡´ì¬í•˜ì§€ ì•ŠëŠ”ë‹¤ë©´
        f_dic[c]=1  #ìƒˆë¡œìš´ ìš”ì†Œ {c:1} ìŒì„ f_dicì— ì¶”ê°€í•œë‹¤. 
  return f_dic

ret1 = getTextfreq('/content/sample_data/exercise/crime_punish_ch_one.txt')
ret2 = sorted(ret1.items(), key=lambda x:x[1], reverse=True) #ret1ì„ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•˜ì—¬ ret2ë¡œ ì¬ì§€ì •. ret2ëŠ” getTextfreq()í•¨ìˆ˜ê°€ ë¦¬í„´í•œ dictionary!!!! ì‚¬ì „ ìë£Œ ì •ë¦¬ëŠ” refer to #131.
#sorted()functionì—ì„œ ì²« ë²ˆì§¸ ì¸ìëŠ” variable.items()module, ì„¸ ë²ˆì§¸ ì¸ìëŠ” ascending or descending, ë‘ ë²ˆì§¸ ì¸ìëŠ” dictionaryì˜ keyê°’ì¸ë° key=lamda x:x[1] ì´í•´X 

for c, freq in ret2:
  if c =='\n': #ë¬¸ìê°€ ì¤„ë°”ê¿ˆ ê¸°í˜¸('\n')ì´ë©´
    continue  #í™”ë©´ì— ì¶œë ¥í•˜ì§€ ì•Šê³  ê·¸ëŒ€ë¡œ ë„˜ì–´ê°„ë‹¤.
  print('[%c] -> [%d]íšŒ ë‚˜íƒ€ë‚¨' %(c,freq))
```

```python
#@markdown ğŸ¯ğŸ“ŒíŒŒì¼ì—ì„œ íŠ¹ì • ë‹¨ì–´ ê°œìˆ˜ ê³„ì‚°í•˜ê¸° #168
#positionì— ê´€ë ¨í•œ 92 ë²ˆ ì˜ˆì œë¥¼ ë” ë³´ê³  defined function ì´í•´í•˜ê¸°

def countWord(filename, word):
  with open(filename, 'rt') as f:
    text = f.read()
    text = text.lower()
    pos = text.find(word)
    count = 0
    while pos!=-1: #ğŸ’ŠRefer to 92ë²ˆ ì˜ˆì œ
      count+=1
      pos = text.find(word,pos+1) #ğŸ’Š ì¸ì ì¤‘ê°„ì— period ê´€ì„±ì ìœ¼ë¡œ ì•ˆì“°ê²Œ ì¡°ì‹¬í•˜ê¸°
  return count

word = input('ê´€ì‹¬ìˆëŠ” í…ìŠ¤íŠ¸ íŒŒì¼ì—ì„œ ì—ì„œ ë‹¨ì–´ ê°œìˆ˜ë¥¼ êµ¬í•  ë‹¨ì–´ë¥¼ ì…ë ¥í•˜ì„¸ìš”:')
word = word.lower()
ret = countWord('/content/sample_data/exercise/crime_punish_ch_one.txt', word) #ğŸ’Š defined functionì— ë”°ë¼ ì¸ìë¥¼ ë„£ì–´ì•¼ì§€, ê´€ì„±ì ìœ¼ë¡œ (íŒŒì¼ë„¤ì„, 'rt')í•˜ë©´ ì•ˆëœë‹¤. 
print('[%s]ì˜ ê°œìˆ˜: [%d]' %(word, ret))
```
